```python
import pandas as pd
import numpy as np
import psycopg2
```

```python
# 读取winemag-data_first150k.csv文件，且索引为其第一列
BaileyAve = pd.read_csv('BaileyAve.csv')
DelawareAve = pd.read_csv('DelawareAve.csv')
ElmwoodAve = pd.read_csv('ElmwoodAve.csv')
FillmoreAve = pd.read_csv('FillmoreAve.csv')
HertelAve = pd.read_csv('HertelAve.csv')
JeffersonAve = pd.read_csv('JeffersonAve.csv')
MainSt = pd.read_csv('MainSt.csv')
NiagaraSt = pd.read_csv('NiagaraSt.csv')
# 保存animals数据集为文件cows_and_goats.csv
#animals.to_csv("cows_and_goats.csv")
```

```python
BaileyAvePT = BaileyAve.pivot_table(index=['Intersection_ID','starttime'], columns=['startdate'], values=['ped'], aggfunc=np.sum)
DelawareAve = DelawareAve.pivot_table(index=['Intersection_ID','starttime'], columns=['startdate'], values=['ped'], aggfunc=np.sum)
ElmwoodAve = ElmwoodAve.pivot_table(index=['Intersection_ID','starttime'], columns=['startdate'], values=['ped'], aggfunc=np.sum)
FillmoreAve = FillmoreAve.pivot_table(index=['Intersection_ID','starttime'], columns=['startdate'], values=['ped'], aggfunc=np.sum)
HertelAve = HertelAve.pivot_table(index=['Intersection_ID','starttime'], columns=['startdate'], values=['ped'], aggfunc=np.sum)
JeffersonAve = JeffersonAve.pivot_table(index=['Intersection_ID','starttime'], columns=['startdate'], values=['ped'], aggfunc=np.sum)
MainSt = MainSt.pivot_table(index=['Intersection_ID','starttime'], columns=['startdate'], values=['ped'], aggfunc=np.sum)
NiagaraSt = NiagaraSt.pivot_table(index=['Intersection_ID','starttime'], columns=['startdate'], values=['ped'], aggfunc=np.sum)
```

```python
BaileyAvePT.to_csv('BaileyAvePT1.csv')
DelawareAve.to_csv('DelawareAve1.csv')
ElmwoodAve.to_csv('ElmwoodAve1.csv')
FillmoreAve.to_csv('FillmoreAve1.csv')
HertelAve.to_csv('HertelAve1.csv')
JeffersonAve.to_csv('JeffersonAve1.csv')
MainSt.to_csv('MainSt1.csv')
NiagaraSt.to_csv('NiagaraSt1.csv')
```
